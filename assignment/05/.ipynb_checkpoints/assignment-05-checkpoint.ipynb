```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

1. Plotting the training data set
    - Load data from 'data.txt' file.
    - Split data into points, labels.
    - Use Scatter() function to plot the training data.
    - We will use a matrix $X$ as a traning data set.

```python
path = "data.txt"
data    = np.genfromtxt(path, delimiter=',')

x       = data[:, 0:1]
y       = data[:, 1:2]
label   = data[:, 2:3]

x_label0    = x[label == 0]
x_label1    = x[label == 1]
y_label0    = y[label == 0]
y_label1    = y[label == 1]

plt.figure(figsize=(8, 8))
plt.scatter(x_label0, y_label0, alpha=0.3, c='b')
plt.scatter(x_label1, y_label1, alpha=0.3, c='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('1. The training data')
plt.show()
```

2. Setting varialbes
    - 'm' is the number of data points.
    - Transpose x, y, label vectors to calculate cost function.
    - 'Learning rate' is 0.0003.
    - Make empty list to store '$\theta$' and plot later.
    - Make empty list to store 'J', the energy function, and plot later
    - 't' means the iteration of gradient descent.
    - Initiate the $\theta_0, \theta_1, \theta_2$, we will use this parameter as a matrix $\theta$

```python
m = len(x)
x_0 = np.ones((1, m), dtype=np.float64)
x_1 = x.T
x_2 = y.T
label = label.T
feature = np.concatenate((x_0, x_1, x_2), axis=0)

learning_rate = 0.0003

theta=np.ones((3,1), dtype=np.float64)
theta[0,0]=-10
theta[1,0]=2
theta[2,0]=1

list_theta_0 = list()
list_theta_1 = list()
list_theta_2 = list()

t = 1
iteration = list()
temp_j_value = 0
j_value = list()
```

3. Gradient descent
    - The cost function is $J(\theta_0, \theta_1, \theta_2) = \frac{-1}{m}\sum_{i=1}^{m}(l^{(i)}log(\sigma(z^{(i)})) + (1 - l^{(i)})log(1 - \sigma(z^{(i)}))) $.
    - '$\sigma(z)$' is sigmoid function. $\sigma(z) = \frac{1}{1+e^{(-z)}}$.
    - $z$ is $\theta^{T}X$.
    - $\theta^{(t+1)}$ := $\theta^{(t)}$ - $\alpha\frac{1}{m}$$\sum_{i=1}^{m}{X (\sigma(\theta^{T}X) - l)}$.
    - Store parameters $\theta$ and cost J to plot later.

```python
while True:
    z = (theta.T).dot(feature)
    sigmoid = 1 / (1 + np.exp(-z))
    if t>1 and abs(temp_j_value - (-(np.sum((label*np.log(sigmoid))+(1-label)*np.log(1-sigmoid)))/m)) <= 0.00001:
        fin_theta = theta
        break
    temp_j_value = -(np.sum((label*np.log(sigmoid))+(1-label)*np.log(1-sigmoid)))/m
    j_value.append(temp_j_value)
    iteration.append(t)
    #print(t, temp_j_value, '\n')
    t += 1

    list_theta_0.append(theta[0,0])
    list_theta_1.append(theta[1,0])
    list_theta_2.append(theta[2,0])
    theta = theta - learning_rate * feature.dot((sigmoid-label).T) / m
```

4. Plotting the estimated parameters
    - This let us know each parameters at every gradient descent iteration.
    - We can know the parameters converge eventually.

```python
plt.figure(figsize=(8, 8))
plt.plot(iteration, list_theta_0, c = 'red')
plt.plot(iteration, list_theta_1, c = 'green')
plt.plot(iteration, list_theta_2, c = 'blue')
plt.xlabel('t : iteration')
plt.ylabel('value')
plt.title('2. The estimated parameters')
plt.legend(['theta0', 'theta1', 'theta2'])
plt.grid()
plt.show()
```

5. Plotting the training error.
    - The energy(cost) converges at certain value which is the optimal of our cost function.

```python
plt.figure(figsize=(8, 8))
plt.plot(iteration, j_value, c = 'blue')
plt.xlabel('t : iteration')
plt.ylabel('cost value')
plt.title('3. The training error')
plt.legend(['J(theta)'])
plt.grid()
plt.show()
```

6. Plotting the obtained classifier.
    - The data points in red area have label 1.
    - The data points in blue area have label 0.
    - At the boundary of two areas, data points can be classified wrong.

```python
xlin = np.arange(30,100,0.5)
ylin = np.arange(30,100,0.5)
X,Y = np.meshgrid(xlin, ylin)
Z = fin_theta[0,0]+fin_theta[1,0]*X+fin_theta[2,0]*Y

plt.figure(figsize=(8, 8))
plt.scatter(x_label0, y_label0, alpha=0.3, c='b')
plt.scatter(x_label1, y_label1, alpha=0.3, c='r')
plt.imshow(Z, cmap='coolwarm', extent=(30,100,30,100), origin='lower')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('4. The obtained classifier')
plt.show()
```
