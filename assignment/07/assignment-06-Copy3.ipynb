```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import warnings#ignore warning messages
warnings.filterwarnings("ignore")

data    = np.genfromtxt("data-nonlinear.txt", delimiter=',')

pointX  = data[:, 0:1]
pointY  = data[:, 1:2]
label   = data[:, 2:3]

pointX0 = pointX[label == 0]
pointY0 = pointY[label == 0]

pointX1 = pointX[label == 1]
pointY1 = pointY[label == 1]
```

```python
#number of the traing data
m = len(pointX)

#zero_term = np.ones((1,m))
x = pointX.T
y = pointY.T

feature = np.ones((1,m))

for i in range(10):
    for j in range(10):
        temp_data = (x**(i))*(y**(j))
        feature = np.concatenate((feature, temp_data), axis=0)
        #print("x", i, "y", j)
        
feature = feature[1:,:]

label = label.T

learning_rate = 0.1
lamda = 0
#initiate theata
theta = np. ones((100, 1))

#lists for storing changes in theta at every iteration

#variable that count iteration
t = 0
theta_list = list()
theta_list.append(theta)
iteration = list()
#list that store j values at every iteration
temp_j_value = 0
j_value = list()
#list for storing accuracy
accuracy_list = list()
#theta[0,0]=20
#theta[1,0]=5
#theta[11,0]=5
#theta[3,0]=10
#theta[21,0]=10
#(1-(learning_rate*lamda)/m)*
```

```python
while True:
    g = (theta.T).dot(feature)
    sigmoid = 1 / (1 + np.exp(-g))
    #if there is no change in j value, then break
    if t>1 and abs(temp_j_value - (-(np.sum((label*np.log(sigmoid))+(1-label)*np.log(1-sigmoid)))/m)) <= 0.000001:
        fin_theta = theta
        break
    temp_j_value = -(np.sum((label*np.log(sigmoid))+(1-label)*np.log(1-sigmoid)))/m
    j_value.append(temp_j_value)
    iteration.append(t)
    print(t, temp_j_value, '\n')
    
    #store at theta list
    #for calculating accuracy
    
    pred0 = (theta.T).dot(feature)
    pred1 = 1 / (1 + np.exp(-pred0))
    
    for i in range(118):
        if (pred1[0, i]>=0.5):
            pred1[0, i]=1
        else:
            pred1[0, i]=0
    correct = 0
    for i in range(118):
        if (pred1[0, i]==label[0,i]):
            correct +=1
    accuracy = correct / 118
    accuracy_list.append(accuracy)
    
    
    #updata theta
    theta_list.append(theta)
    theta = (1-(learning_rate*lamda)/m)*theta - learning_rate * feature.dot((sigmoid-label).T) / m
    t+=1
```

```python
#theses are used for plotting
xlin = np.arange(-2, 2, 0.01)
ylin = np.arange(-2, 2, 0.01)
X, Y = np.meshgrid(xlin, ylin)
theta_num = 0
Z = np.zeros((400,400))
for i in range(10):
    for j in range(10):
        Z += theta[theta_num, 0]*((X**(i))*(Y**(j)))
        theta_num += 1
sigmoids = 1 / (1 + np.exp(-Z))
```

2. Writing down the high dimensional function $g(x, y; \theta)$
    - k = 15
    - $g(x, y; \theta)$ = $2.697 + 2.342x + 3.363y - 3.621xy - 3.190x^2 - 2.437y^2 - 0.033x^3 - 2.378x^2y - 2.144xy^2 - 1.523y^3 - 7.305x^4 + 0.671x^3y - 3.096x^2y^2 - 1.007xy^3 - 6.210y^4$

- Gradient descent
    - The cost function is 
    $J(\theta)$ = $\frac{-1}{m}\sum_{i=1}^{m}(l^{(i)}log(\sigma(z^{(i)}))$ + $(1 - l^{(i)})log(1 -\sigma(z^{(i)}))) $.
    - '$\sigma(z)$' is sigmoid function. $\sigma(z) = \frac{1}{1+e^{(-z)}}$.
    - $z$ is $\theta^{T}X$.
    - $l$ is label.
    - $\theta^{(t+1)}$ := $\theta^{(t)}$ - $\alpha\frac{1}{m}$$\sum_{i=1}^{m}{X (\sigma(\theta^{T}X) - l)}$.
    - Store parameters $\theta$ and cost J to plot later.
